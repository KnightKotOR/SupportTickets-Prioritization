Ноутбук для создания сэмплов набора данных для тестирования

```python
from pathlib import Path

import pandas as pd
```

```python
dataset_path = Path("../data/Support_tickets.csv")
train_fixture_path = Path("./data/fixtures/train_sample.csv")

df = pd.read_csv(dataset_path, nrows=7000)
df = df.select_dtypes(include=["number"])
df.drop(columns=["ticket_id", "org_users"], inplace=True)

df.to_csv(train_fixture_path, index=False)
```

```python
test_fixture_path = Path("./data/fixtures/test_sample.csv")

test_df = pd.read_csv(dataset_path, nrows=3000, skiprows=lambda x: x in range(1, 7001))
test_df = test_df.select_dtypes(include=["number"])
test_df.drop(columns=["ticket_id", "org_users"], inplace=True)

test_df.to_csv(test_fixture_path, index=False)
```
